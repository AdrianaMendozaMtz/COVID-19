```python
import pandas as pd

import matplotlib as mpl


%matplotlib inline
pd.set_option('display.max_columns', 68)
```

```python
url='https://covid.ourworldindata.org/data/owid-covid-data.csv'
```

```python
pd_raw=pd.read_csv(url,sep=',')
pd_raw
```

```python
pd_raw.to_csv('Covid_19_cases_data.csv',';')
```

```python
pd_MEX=pd_raw[pd_raw['location']=='Mexico']
pd_MEX
```

```python
pd_MEX=pd_MEX.reset_index(drop=True)
pd_MEX
```

```python
import plotly.express as px 


fig=px.line(pd_MEX,x="date",y="people_fully_vaccinated", title='Vaccination rate COVID-19')
fig.show()
```

```python
b_MEX=pd_raw['location']=='Mexico'
b_UK=pd_raw['location']=='United Kingdom'
b_US=pd_raw['location']=='United States'

```

```python
b_all=b_MEX|b_UK|b_US
```

```python
pd_list=pd_raw[b_all]
pd_list
```

```python
pd_list['people_fully_vaccinated/population']=pd_list['people_fully_vaccinated']/pd_list['population']
```

```python
fig=px.line(pd_list,x="date",y="people_fully_vaccinated/population", color='location',title='Vaccination rate COVID-19')
fig.show()
```
